```python
# --- テスト ---
class TestDistributedID3(unittest.TestCase):
    def setUp(self):
        # テスト用の3分割データ
        full_df = pd.DataFrame({
            'Outlook': ['Sunny', 'Sunny', 'Overcast', 'Rain', 'Rain', 'Rain', 'Overcast', 'Sunny', 'Sunny', 'Rain', 'Sunny', 'Overcast', 'Overcast', 'Rain'],
            'Temperature': ['Hot', 'Hot', 'Hot', 'Mild', 'Cool', 'Cool', 'Cool', 'Mild', 'Cool', 'Mild', 'Mild', 'Mild', 'Hot', 'Mild'],
            'Humidity': ['High', 'High', 'High', 'High', 'Normal', 'Normal', 'Normal', 'High', 'Normal', 'Normal', 'Normal', 'High', 'Normal', 'High'],
            'Wind': ['Weak', 'Strong', 'Weak', 'Weak', 'Weak', 'Strong', 'Strong', 'Weak', 'Weak', 'Weak', 'Strong', 'Strong', 'Weak', 'Strong'],
            'PlayTennis': ['No', 'No', 'Yes', 'Yes', 'Yes', 'No', 'Yes', 'No', 'Yes', 'Yes', 'Yes', 'Yes', 'Yes', 'No']
        })
        
        self.party_a = Site("A社", full_df[['Outlook']])
        self.party_b = Site("B社", full_df[['Temperature', 'Humidity']])
        self.party_c = Site("C社", full_df[['Wind', 'PlayTennis']])

    def test_get_split_indices(self):
        # IDが [0, 1, 2, 3] の4件だけが対象だとする
        target_ids = [0, 1, 2, 3]
        
        # Party A に Outlook で分割してもらう
        split_result = self.party_a.get_split_indices('Outlook', target_ids)
        
        # 期待される結果:
        # ID 0, 1 は 'Sunny'
        # ID 2 は 'Overcast'
        # ID 3 は 'Rain'
        # （ID 4, 5 は target_ids に含まれていないので無視されること）
        expected = {
            'Sunny': [0, 1],
            'Overcast': [2],
            'Rain': [3]
        }
        
        self.assertEqual(split_result, expected)

if __name__ == '__main__':
    unittest.main(argv=['first-arg-is-ignored'], exit=False)
```
